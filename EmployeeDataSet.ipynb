## Removing Outliers

```python
import pandas as pd
import numpy as np
import seaborn as sns
df=pd.read_excel('employee.xlsx')
df
```

```python
sns.boxplot(df.Salary)
```

```python
df.fillna({
    "Department":df.Department.mode()[0],
    "Salary":df.Salary.mean(),
    "Age":df.Age.median()},inplace=True)
df
```

# Finding lower and upper fence

```python
def outlier_finder(s):
    q1=np.percentile(s,25)
    q3=np.percentile(s,75)
    iqr=q3-q1
    lf=q1 - 1.5 * iqr
    uf=q3 + 1.5 * iqr
    return lf, uf
lf,uf=outlier_finder(df.Salary)
print(f"Lower fence {lf}")
print(f"Upper fence {uf}")
df
```

```python
outlier=df[(df.Salary<lf)|(df.Salary>uf)]
outlier
```

```python
sns.boxplot(df.Salary)
```

```python
df.describe()
```

```python
#saving data in new variable with outlier
df2=df[(df.Salary>=lf)&(df.Salary<=uf)]
df2
```

```python
# 1. Add a new Column Revised_Salary, revised salary is 10% more than actual salary
df["Revised_Salary"]=df.Salary+(10/100)*df.Salary
```

```python
df
```

```python
# 2. Calculate total additional cost to company
add_cost=df.Revised_Salary.sum()
print(add_cost)
```

```python
add_cost_dep=df.groupby("Department")["Revised_Salary"].sum()
add_cost_dep
```

```python
# 3. Which Department faces the biggest impact
biggest_impact=df.groupby("Department")["Revised_Salary"].sum()-df.groupby("Department")["Salary"].sum()
```

```python
biggest_impact_id = biggest_impact.idxmax()
biggest_impact_value = biggest_impact.max()
print(f"The department facing the biggest impact is ({biggest_impact_id}) with an additional cost of ({biggest_impact_value})")
```
